```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/heart-disease-cleveland-uci/heart_cleveland_upload.csv", sep="," )
```

```python
X = df.iloc[:,0:13]
Y = df.iloc[:,13]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X, Y, test_size=0.3, random_state=1)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(8, input_shape=(13,), activation=tf.nn.relu))
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)) 
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
history = model.fit(x=x_train, y=y_train, epochs=150,
                    validation_data=(x_test, y_test), verbose=0)
```

```python
val_loss, val_acc = model.evaluate(x=x_test, y=y_test)
```

```python
themodel = "./heart_ann.h5"
model.save(themodel)
```
